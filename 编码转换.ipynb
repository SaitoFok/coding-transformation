## 最新处理方法


```python
from docx import Document
from docx.shared import RGBColor

def unicode_mapping(text, num):
    result = []
    for char in text:
        unicode_value = ord(char)
        mapped_char = chr(unicode_value - num)
        result.append(mapped_char)
    return "".join(result)

def apply_color_and_bold(run, color, is_bold):
    if color:
        run.font.color.rgb = color
    if is_bold:
        run.font.bold = True

def read_and_transform_docx(input_path, output_path, num):
    doc = Document(input_path)

    # Create a new document for the transformed text
    transformed_doc = Document()

    for paragraph in doc.paragraphs:
        original_text = ""
        original_color = None
        original_bold = False

        # Concatenate runs in the same paragraph
        for run in paragraph.runs:
            original_text += run.text
            original_color = run.font.color.rgb if run.font.color else None
            original_bold = run.font.bold or original_bold

        transformed_text = unicode_mapping(original_text, num)
        print(transformed_text)
        # Check if the transformed text contains 【 or 】
        if "【" in transformed_text or "】" in transformed_text:
            # Check if the transformed text contains "校記"
            if "校記" in transformed_text:
                # Replace the entire line with "【校記】"
                transformed_text = "【校記】"
                new_run = transformed_doc.add_paragraph().add_run(transformed_text)
                apply_color_and_bold(new_run, original_color, original_bold)
            # If the line contains 【 or 】, skip writing to the new document
        else:
            # Create a new run with the transformed text and apply the original color and bold style
            new_run = transformed_doc.add_paragraph().add_run(transformed_text)
            apply_color_and_bold(new_run, original_color, original_bold)

    # Save the transformed document
    transformed_doc.save(output_path)

if __name__ == "__main__":
    input_docx_path = "./诗集/诗续集卷下.docx"  # 替换为你的输入.docx文件路径
    output_docx_path = "./诗集（修改）/诗续集卷下_2.docx"  # 替换为输出的.docx文件路径
    num = 4
    read_and_transform_docx(input_docx_path, output_docx_path, num)
    print(f"转化后的文本已保存至: {output_docx_path}")
```
